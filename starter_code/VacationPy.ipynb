```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import gkey
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
city_data = pd.read_csv("output_data/clean_city_data.csv", dtype="object")

# Visualize
city_data = pd.DataFrame(city_data)
city_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps with API key
gmaps.configure(api_key=gkey)
```

```python
locations = city_data[["latitude", "longitude"]].astype(float)
weight = city_data['humidity']
```

```python
fig = gmaps.figure(center=(0,0), zoom_level=2)

heat_layer = gmaps.heatmap_layer(locations, weights=weight, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
city_data['temp'] = city_data['temp'].astype(float)
city_data['cloudiness'] = city_data['cloudiness'].astype(float)
city_data['wind speed'] = city_data['wind speed'].astype(float)
```

```python
Perfect_Weather = city_data.loc[(city_data['temp'] >=68) & (city_data['temp'] <= 78) &
                                (city_data['cloudiness'] <=5 ) & 
                                (city_data['wind speed'] <= 10)]

Perfect_Weather
```

```python
Perfect_Weather["Hotel Name"] = ""
Perfect_Weather.head()
```

```python
# create a params dict that will be updated with new city each iteration
params = {"radius": 5000,
          "types": "hotel",
          "keyword": "hotel",
          "key": gkey}

for index, row in Perfect_Weather.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json" 
    
    # Use the search term: "Hotel and and lat/lng
   
    # get lat, lng from df
    lat = row["latitude"]
    lng = row["longitude"]
    
    params["location"] = f"{lat},{lng}"   

    # make request and print url
    hotel_data = (requests.get(base_url, params=params)).json()
    
    try:
        print(hotel_data["results"][0]["name"]) 
        Perfect_Weather.loc[index, "Hotel Name"] = hotel_data["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing")
```

```python
Hotel_list = Perfect_Weather['Hotel Name'].replace('', np.nan, inplace=True)
Hotel_list = Hotel_list.dropna(subset=['Hotel Name'], inplace=True)
Hotel_list
```

```python
locations = Perfect_Weather[["latitude", "longitude"]].astype(float)
```

```python
markers = gmaps.marker_layer(locations)

fig = gmaps.figure(center=(0,0), zoom_level=2)
fig.add_layer(heat_layer)
fig.add_layer(markers) 
fig
plt.savefig('Map: Hotel Locations.png')
fig
```
